```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# 1. Load the CSV file
df = pd.read_csv('fer2013.csv')

def prepare_data(data):
    """Converts pixel strings in CSV to 4D numpy arrays"""
    pixels = data['pixels'].tolist()
    width, height = 48, 48
    faces = []
    for pixel_sequence in pixels:
        face = [int(pixel) for pixel in pixel_sequence.split(' ')]
        face = np.asarray(face).reshape(width, height)
        faces.append(face.astype('float32') / 255.0)

    faces = np.asarray(faces)
    faces = np.expand_dims(faces, -1) # Add grayscale channel
    return faces

# 2. Split into Train and Test based on the 'Usage' column
train_data = df[df['Usage'] == 'Training']
test_data = df[df['Usage'] == 'PublicTest']

X_train = prepare_data(train_data)
y_train = tf.keras.utils.to_categorical(train_data['emotion'], num_classes=7)

X_test = prepare_data(test_data)
y_test = tf.keras.utils.to_categorical(test_data['emotion'], num_classes=7)

# 3. Build and Train the Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(7, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("Starting training...")
model.fit(X_train, y_train, batch_size=64, epochs=20, validation_data=(X_test, y_test))

# 4. Save the Brain
model.save('my_emotion_model.h5')
print("Model saved as my_emotion_model.h5")
```

```python
# @title Default title text
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.utils import to_categorical

# Load the file
df = pd.read_csv('fer2013.csv')

# Prepare X (images) and y (labels)
X_train = prepare_data(df[df['Usage'] == 'Training'])
y_train = to_categorical(df[df['Usage'] == 'Training']['emotion'], 7)

# Build a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(7, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Start learning
model.fit(X_train, y_train, epochs=10, batch_size=64)
model.save('my_emotion_model.h5')
```

```python
import cv2
import numpy as np
from keras.models import load_model

# 1. Load the brain and the face locator
model = load_model('my_emotion_model.h5')
face_haar = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# The categories must stay in this exact order
labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret: break

    # Convert to grayscale (the model only learned from gray faces)
    gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_haar.detectMultiScale(gray_img, 1.3, 5)

    for (x, y, w, h) in faces:
        # Draw a box around your face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # Crop the face and resize it to 48x48
        roi_gray = gray_img[y:y+h, x:x+w]
        roi_gray = cv2.resize(roi_gray, (48, 48))

        # Prepare the image for the model
        img_pixels = roi_gray.astype('float32') / 255.0
        img_pixels = np.expand_dims(img_pixels, axis=0) # Add batch dimension
        img_pixels = np.expand_dims(img_pixels, axis=-1) # Add channel dimension

        # ANALYZE: Compare live face to the trained dataset
        predictions = model.predict(img_pixels)
        max_index = np.argmax(predictions[0]) # Find the highest probability
        emotion = labels[max_index]

        # Display the result on your screen
        cv2.putText(frame, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Real-time Emotion Analysis', frame)

    # Press 'q' to stop the webcam
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
